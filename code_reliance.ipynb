```python
import ta
```

####  1. SMA(Simple Moving Average)

Calculates the average closing price over 20 trading days.

```python
reliance_df['sma_20'] = ta.trend.sma_indicator(reliance_df['close'], window=20)
```

#### 2. EMA(Exponential Moving Average)
Gives more importance to recent prices than SMA.

```python
reliance_df['ema_20'] = ta.trend.ema_indicator(reliance_df['close'], window=20)
```

#### 3. RSI(Relative Strength Index)
Measures momentum of price movements.

- RSI > 70 - Overbought
- RSI < 30 - Oversold

```python
reliance_df['rsi'] = ta.momentum.rsi(reliance_df['close'], window=14)
```

#### 4. ROC(Rate of Change)
Measures percentage price change over a period.


```python
reliance_df['roc'] = ta.momentum.roc(reliance_df['close'], window=12)
```

#### 5. Bollinger Bands
Measures market volatility.
- Upper Band = SMA + 2sigma
- Lower Band = SMA -2sigma

```python
bb = ta.volatility.BollingerBands(
    reliance_df['close'], 
    window=20, 
    window_dev=2)
reliance_df['bb_high'] = bb.bollinger_hband()
reliance_df['bb_low'] = bb.bollinger_lband()
```

#### 6. ATR(Average True Range)
Measures market volatility.
High ATR indicates larger daily price movements.


```python
reliance_df['atr'] = ta.volatility.average_true_range(
    reliance_df['high'], 
    reliance_df['low'], 
    reliance_df['close'], 
    window=14
    )
```

#### 7. MACD(Moving Average Convergence Divergence)
- MACD = EMA12 - EMA26
- Signal Line - EMA9(MACD)

Shows trend direction and momentum

- MACD above Signal → Bullish
- MACD below Signal → Bearish

```python
reliance_df['macd'] = ta.trend.macd(
    reliance_df['close']
)

reliance_df['macd_signal'] = ta.trend.macd_signal(
    reliance_df['close']
)
```

#### 8. OBV(On Balance Volume)
Combines vloume and price movement.

        OBVprev + Volume Close > Previous close
        OBVprev - Volume Close < Previous close

```python
reliance_df['obv']=ta.volume.on_balance_volume(
    reliance_df['close'], 
    reliance_df['volume']
    )
```

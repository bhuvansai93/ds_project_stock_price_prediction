### Cleaned Datasets into new folder


```python
companys =["Reliance","Infosys", "HDFC_Bank", "Sun_Pharma", "Maruti"] 

for company in companys:
    df = pd.read_csv(f"Datasets/{company}.csv")
    df = df.iloc[2:].reset_index(drop=True)
    df.rename(columns={"Price": "Date"}, inplace=True)
    df.columns = df.columns.str.lower()
    df.to_csv(f"cleaned_datasets/{company}.csv", index=False)
    print(df.shape)
    
```

### 2. Closing Price Trend

```python
plt.figure(figsize=(12, 6))

# Line plot
sns.lineplot(data=reliance_df, x=reliance_df.index, y='close', linewidth=1, label='Close Price')

# Find highest and lowest close price
max_price = reliance_df['close'].max()
min_price = reliance_df['close'].min()

max_date = reliance_df['close'].idxmax()
min_date = reliance_df['close'].idxmin()

# Mark points
plt.scatter(max_date, max_price, s=100, marker='^', c='darkgreen', label=f'High: ₹{max_price:.2f}')
plt.scatter(min_date, min_price, s=100, marker='v', c='red', label=f'Low: ₹{min_price:.2f}')

# Add value labels
plt.annotate(f'₹{max_price:.2f}',
             xy=(max_date, max_price),
             xytext=(2, 5),
             textcoords='offset points')

plt.annotate(f'₹{min_price:.2f}',
             xy=(min_date, min_price),
             xytext=(10, -10),
             textcoords='offset points')

plt.title('Reliance Closing Price Over Time')
plt.xlabel('Year')
plt.ylabel('Close Price (INR)')
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```

### 3. Daily Trading Range

```python
reliance_df['range'] = reliance_df['high'] - reliance_df['low']

plt.figure(figsize=(12,6))
plt.plot(reliance_df.index, reliance_df['range'])
plt.title('Daily Trading Range')
plt.xlabel('Year')
plt.ylabel('Price Difference (INR)')
plt.grid(alpha=0.3)
plt.show()
```

- Not much price difference highest is 137.73

### 4. Trading Volume Trend

```python
from matplotlib.ticker import FuncFormatter

plt.figure(figsize=(12, 6))
sns.lineplot(data=reliance_df, x=reliance_df.index, y='volume', label='Trading Volume')
plt.title('Reliance Trading Volume Over Time')
plt.xlabel('Year') 

plt.gca().yaxis.set_major_formatter(
    FuncFormatter(lambda x, _: f'{x/1e6:.0f}M')
)
plt.ylabel('Trading Volume (Million Shares)')
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```

- Exponential Notation (1e6)= 1* 10^6 = 1Million
- Trading Volume is the total number of shares bought and sold during a trading day.
- Highest Trading Volume is 142.68M

### 7. Daily Return Distribution

```python
reliance_df['daily_return'] = reliance_df['close'].pct_change()*100
```

```python
reliance_df.head()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(reliance_df['daily_return'].dropna(),
             bins=30,
             kde=True)
plt.title('Distribution of Daily Returns')
plt.xlabel('Daily Return (%)')
plt.ylabel('Frequency')
plt.grid(alpha=0.3)
plt.show()
```

### 8. Daily Return Trend

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=reliance_df, x=reliance_df.index, y='daily_return', linewidth=1, label='Daily Return')
plt.title('Reliance Daily Returns Over Time')
plt.xlabel('Year')
plt.ylabel('Daily Return (%)')
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```

### 9. Correlation Heatmap

```python
plt.figure(figsize=(12, 6))
corr = reliance_df[['open','high',
           'low','close',
           'volume']].corr()

sns.heatmap(corr,
            annot=True,
            cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

- Prices move together
- Lower stock prices were often associated with higher trading volumes.

### 10. Candlestick Chart

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Candlestick(
    x=reliance_df.index,
    open=reliance_df['open'],
    high=reliance_df['high'],
    low=reliance_df['low'],
    close=reliance_df['close']
)])

fig.update_layout(
    title='Reliance Candlestick Chart',
    xaxis_title='Date',
    yaxis_title='Price (INR)',
    xaxis_rangeslider_visible=False
)

fig.show()
```
